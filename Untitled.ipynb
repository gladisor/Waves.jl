```julia
using DifferentialEquations
using Plots

# Define the differential equation function
function simple_ode(du, u, p, t)
    # Define the ODE: du/dt = -k * u
    k = p
    du[1] = -k * u[1]
end

# Initial condition
u0 = [1.0]

# Time span
tspan = (0.0, 10.0)

# Parameters
k = 1.0

# Define the ODE problem
prob = ODEProblem(simple_ode, u0, tspan, k)

# Solve the ODE
sol = solve(prob)

# Plot the solution
plot(sol, xlabel="Time", ylabel="Solution", label="u(t)")
```
